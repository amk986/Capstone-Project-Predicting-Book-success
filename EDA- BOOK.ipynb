There are 30 files broken up into 10M user data on books obtained from goodreads. Concatenate the exel files to put into one dataframe.


```python
#empty dataframe for columns

goodreads_10M_books = pd.DataFrame(columns = pd.read_csv(r"C:\Users\alido\OneDrive\Desktop\goodreads_10M\book1000k-1100k.csv", index_col = "Id").columns)
goodreads_10M_books
```

```python
for dirname, _, filenames in os.walk(r"C:\Users\alido\OneDrive\Desktop\goodreads_10M"):
    for filename in filenames:
        if "_" not in filename:
            goodreads_10M_books = pd.concat([goodreads_10M_books, pd.read_csv(os.path.join(dirname, filename), index_col = 'Id')])
            print(os.path.join(dirname, filename), 'OK')
```
